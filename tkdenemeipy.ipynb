```python
import tkinter as tk
from tkinter import Canvas

def draw_cylinder(canvas, x, y, height, radius):
    # Draw the main cylinder body
    canvas.create_polygon(x - radius, y, x + radius, y, x + radius * 0.8, y + height, x - radius * 0.8, y + height, fill="gray", outline="")
    
    # Draw the ellipse bases of the cylinder
    canvas.create_arc(x - radius, y, x + radius, y + height * 2, start=0, extent=180, fill="gray", outline="")
    canvas.create_arc(x - radius, y, x + radius, y + height * 2, start=180, extent=180, fill="", outline="gray")

def draw_cone(canvas, x, y, height, radius):
    # Draw the main cone body
    canvas.create_polygon(x - radius, y + height, x + radius, y + height, x, y, fill="gray", outline="")
    
    # Draw the ellipse base of the cone
    canvas.create_arc(x - radius, y + height, x + radius, y + height * 2, start=0, extent=180, fill="gray", outline="")

# Create the main application window
root = tk.Tk()
root.title("3D Axes")

# Create a Canvas widget
canvas = Canvas(root, width=400, height=400)
canvas.pack()

# Draw x, y, and z axes with cylinders and cones
# X-axis
draw_cylinder(canvas, 100, 200, 20, 5)
draw_cone(canvas, 100, 180, 20, 10)
# Y-axis
draw_cylinder(canvas, 200, 100, 20, 5)
draw_cone(canvas, 200, 80, 20, 10)
# Z-axis
draw_cylinder(canvas, 300, 200, 20, 5)
draw_cone(canvas, 300, 180, 20, 10)

# Run the application
root.mainloop()
```

```python
import tkinter as tk
from tkinter import Canvas

def draw_cylinder(canvas, x, y, height, radius, orientation):
    # Draw the cylinder
    if orientation == "x":
        canvas.create_rectangle(x, y - radius, x + height, y + radius, fill="gray", outline="")
    elif orientation == "y":
        canvas.create_rectangle(x - radius, y, x + radius, y + height, fill="gray", outline="")
    elif orientation == "z":
        canvas.create_rectangle(x - radius, y - radius, x + radius, y + height, fill="gray", outline="")

# Create the main application window
root = tk.Tk()
root.title("3 Perpendicular Cylinders")

# Create a Canvas widget
canvas = Canvas(root, width=400, height=400)
canvas.pack()

# Define cylinder parameters
cylinder_height = 100
cylinder_radius = 20

# Coordinates of the point where cylinders intersect
intersection_point = (200, 200)

# Draw cylinders for x, y, and z axes
draw_cylinder(canvas, intersection_point[0], intersection_point[1], cylinder_height, cylinder_radius, "x")
draw_cylinder(canvas, intersection_point[0], intersection_point[1], cylinder_height, cylinder_radius, "y")
draw_cylinder(canvas, intersection_point[0], intersection_point[1], cylinder_height, cylinder_radius, "z")

# Run the application
root.mainloop()
```

```python
import tkinter as tk
import math

def draw_cylinder(canvas, x, y, height, radius):
    num_segments = 20  # Number of segments to approximate the circular base
    
    # Calculate points for the circular base
    base_points = []
    for i in range(num_segments):
        angle = 2 * math.pi * i / num_segments
        base_x = x + radius * math.cos(angle)
        base_y = y + radius * math.sin(angle)
        base_points.append((base_x, base_y))
    
    # Draw the circular base
    canvas.create_polygon(base_points, fill="gray", outline="")

    # Draw lines connecting the base points to form the sides of the cylinder
    for i in range(num_segments):
        next_i = (i + 1) % num_segments
        canvas.create_line(base_points[i][0], base_points[i][1], base_points[next_i][0], base_points[next_i][1])
        canvas.create_line(base_points[i][0], base_points[i][1], base_points[i][0], base_points[i][1] + height)
        canvas.create_line(base_points[i][0], base_points[i][1] + height, base_points[next_i][0], base_points[next_i][1] + height)

# Create the main application window
root = tk.Tk()
root.title("Pseudo-3D Cylinder")

# Create a Canvas widget
canvas = tk.Canvas(root, width=400, height=400)
canvas.pack()

# Define cylinder parameters
cylinder_height = 100
cylinder_radius = 20

# Coordinates of the point where cylinders intersect
intersection_point = (200, 200)

# Draw cylinders for x, y, and z axes
draw_cylinder(canvas, intersection_point[0] - cylinder_height // 2, intersection_point[1], cylinder_height, cylinder_radius)
draw_cylinder(canvas, intersection_point[0], intersection_point[1] - cylinder_height // 2, cylinder_height, cylinder_radius)
draw_cylinder(canvas, intersection_point[0] + cylinder_height // 2, intersection_point[1], cylinder_height, cylinder_radius)

# Run the application
root.mainloop()
```

```python
import tkinter as tk
import math

class Cylinder3D(tk.Canvas):
    def __init__(self, master=None, *args, **kwargs):
        tk.Canvas.__init__(self, master, *args, **kwargs)
        self.width = self.winfo_reqwidth()
        self.height = self.winfo_reqheight()
        
    def draw_cylinder(self, x, y, radius, height):
        # Drawing top and bottom circles
        self.create_oval(x - radius, y - radius, x + radius, y + radius, outline="black")
        self.create_oval(x - radius, y + height - radius, x + radius, y + height + radius, outline="black")
        
        # Drawing lines connecting top and bottom circles
        self.create_line(x - radius, y, x - radius, y + height, fill="black")
        self.create_line(x + radius, y, x + radius, y + height, fill="black")
        
        # Drawing curved lines for cylinder
        for i in range(0, 360, 10):
            x1 = x + radius * math.cos(math.radians(i))
            y1 = y + radius * math.sin(math.radians(i))
            x2 = x + radius * math.cos(math.radians(i + 10))
            y2 = y + radius * math.sin(math.radians(i + 10))
            self.create_line(x1, y1, x2, y2, fill="black")

def main():
    root = tk.Tk()
    root.title("3D Cylinders")
    
    canvas = Cylinder3D(root, width=400, height=300)
    canvas.pack(side=tk.LEFT, padx=10)
    
    # Drawing cylinder at specified position and dimensions
    canvas.draw_cylinder(100, 150, 50, 150)
    
    # Drawing another cylinder perpendicular to the first one
    canvas2 = Cylinder3D(root, width=400, height=300)
    canvas2.pack(side=tk.RIGHT, padx=10)
    canvas2.draw_cylinder(30, 150, 50, 150)
    
    root.mainloop()

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
import math

class Cylinder3D(tk.Canvas):
    def __init__(self, master=None, *args, **kwargs):
        tk.Canvas.__init__(self, master, *args, **kwargs)
        self.width = self.winfo_reqwidth()
        self.height = self.winfo_reqheight()
        
    def draw_cylinder(self, x, y, radius, height, rotation_angle=0):
        # Drawing top and bottom circles
        self.create_oval(x - radius, y - radius, x + radius, y + radius, outline="black")
        self.create_oval(x - radius, y + height - radius, x + radius, y + height + radius, outline="black")
        
        # Drawing lines connecting top and bottom circles
        self.create_line(x - radius, y, x - radius, y + height, fill="black")
        self.create_line(x + radius, y, x + radius, y + height, fill="black")
        
        # Drawing curved lines for cylinder
        for i in range(0, 360, 10):
            x1 = x + radius * math.cos(math.radians(i + rotation_angle))
            y1 = y + radius * math.sin(math.radians(i + rotation_angle))
            x2 = x + radius * math.cos(math.radians(i + 10 + rotation_angle))
            y2 = y + radius * math.sin(math.radians(i + 10 + rotation_angle))
            self.create_line(x1, y1, x2, y2, fill="black")

def main():
    root = tk.Tk()
    root.title("3D Cylinders")
    
    canvas = Cylinder3D(root, width=400, height=300)
    canvas.pack(side=tk.LEFT, padx=10)
    
    # Drawing cylinder at specified position, dimensions, and rotation angle
    canvas.draw_cylinder(100, 150, 50, 150, rotation_angle=300)
    
    # Drawing another cylinder perpendicular to the first one
    canvas2 = Cylinder3D(root, width=400, height=300)
    canvas2.pack(side=tk.RIGHT, padx=10)
    canvas2.draw_cylinder(300, 150, 50, 150, rotation_angle=60)
    
    root.mainloop()

if __name__ == "__main__":
    main()
```

```python
import tkinter as tk
import math

# Define the dimensions of the canvas
canvas_width = 400
canvas_height = 400

# Create a Tkinter window
root = tk.Tk()
root.title("3D Axes")

# Create a canvas to draw on
canvas = tk.Canvas(root, width=canvas_width, height=canvas_height, bg="white")
canvas.pack()

# Define initial angles
angle_x = 0
angle_y = 0
angle_z = 0

# Function to draw 3D axes
def draw_axes():
    # Draw x-axis
    canvas.create_line(100, 300, 300, 300, fill="red", arrow=tk.LAST)
    canvas.create_text(320, 300, text="X", fill="red")
    # Draw y-axis
    canvas.create_line(100, 300, 100, 100, fill="green", arrow=tk.LAST)
    canvas.create_text(100, 80, text="Y", fill="green")
    # Draw z-axis
    canvas.create_line(100, 300, 200, 200, fill="blue", arrow=tk.LAST)
    canvas.create_text(220, 180, text="Z", fill="blue")

# Function to rotate axes around z-axis
def rotate_axes():
    global angle_z
    angle_z += 5
    redraw_axes()

# Function to redraw the axes after rotation
def redraw_axes():
    canvas.delete("all")
    draw_axes()
    # Rotate the axes around z-axis
    draw_rotated_axes(angle_z)

# Function to draw rotated axes
def draw_rotated_axes(angle):
    rad_angle = math.radians(angle)
    # Rotate x-axis
    x1 = 100 * math.cos(rad_angle)
    y1 = 300 - 100 * math.sin(rad_angle)
    x2 = 300 * math.cos(rad_angle)
    y2 = 300 - 300 * math.sin(rad_angle)
    canvas.create_line(x1, y1, x2, y2, fill="red", arrow=tk.LAST)
    canvas.create_text(x2 + 20, y2, text="X", fill="red")
    # Rotate y-axis
    x1 = 100 * math.cos(rad_angle + math.pi/2)
    y1 = 300 - 100 * math.sin(rad_angle + math.pi/2)
    x2 = 100 * math.cos(rad_angle + math.pi/2)
    y2 = 300 - 300 * math.sin(rad_angle + math.pi/2)
    canvas.create_line(x1, y1, x2, y2, fill="green", arrow=tk.LAST)
    canvas.create_text(x2, y2 - 20, text="Y", fill="green")
    # Draw z-axis
    canvas.create_line(100, 300, 200, 200, fill="blue", arrow=tk.LAST)
    canvas.create_text(220, 180, text="Z", fill="blue")

# Call the function to draw the initial axes
draw_axes()

# Bind 'a' key press event to rotate the axes around z-axis
root.bind('<a>', lambda event: rotate_axes())

# Run the Tkinter event loop
root.mainloop()
```
